```python
import beam as bm


import numpy as np
import matplotlib.pyplot as plt
import os    
os.environ['KMP_DUPLICATE_LIB_OK']='True' 

from tqdm.notebook import tqdm
```

```python
bm.set_operation_mode('fn')
```

__Multi run with beta sampling__

```python
SENT_VALUE=0.2

mkt=bm.Market()
mkt.add_partition(1000, 'fixed_sentiment')
mkt.set_part_sentiment('fixed_sentiment', bm.sentiment_start_at_value(SENT_VALUE))
mkt.add_metric('sigma_scatter', bm.mtr_sct_sigma)
```

```python
lab=bm.Lab(mkt)
```

```python
lab.multi_run(365, 20)
```

```python
plt.figure(figsize=(15,4))
lab.plot_multi_sim_metric('avg_price')
plt.title('Avgerage daily transaction price (Beta sampling)')
plt.xlabel('time $t$')
plt.ylabel('$<\pi_t>$')
plt.grid()
plt.show()
```
